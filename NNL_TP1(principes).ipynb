```python
def lin(x):
    return x
    
def satlin(x):
    if x>= 0 :
        return x
    return 0

def logsig(x):
    return 1/(1+np.exp(x))

# def summation(w,p,b):
#     return w@p+b

def neuron_output(w,p,b, trans_func):
    return trans_func(w@p+b)
```

```python
p = np.array([0.5, 1.2, -0.7, 2, 1.8, -2])
w = np.array([0.3, -0.6, 0.8, 0.7, 0.9, 0.3])
b = 0.8
neuron_output(w,p,b,lin)
```
